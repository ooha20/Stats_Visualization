## Dropping the missing values in Business data

```python
# Dropping the columns which are more than 180000 missing values using dropna
business_data.dropna(thresh=len(business_data) - 180000, axis=1, inplace=True)
#business_data.info()
# Fill missing values with mode for columns with missing values
business_data.fillna(business_data.mode().iloc[0], inplace=True)
print(business_data.isnull().sum())
```

The output shows the count of missing values for each column in the Business Data.

As columns in our DataFrame has missing values so I just dropped the columns using dropna. Remaining columns I filled with mode.

## Filling missing values with mean and mode in Review data

```python
#Filling the missing values with the mean value and mode
review_data['funny'].fillna(review_data['funny'].mean(), inplace=True)
review_data['review_id'].fillna(review_data['review_id'].mode()[0], inplace=True)
review_data['stars'].fillna(review_data['stars'].mode()[0], inplace=True)
review_data['text'].fillna(review_data['text'].mode()[0], inplace=True)
review_data['useful'].fillna(review_data['useful'].mode()[0], inplace=True)
review_data['user_id'].fillna(review_data['user_id'].mode()[0], inplace=True)
print(review_data.isnull().sum())
```

Filled the missing values with mode and funny column with mean.

## Filling the missing values with mode in User data

```python
#Filling the missing values with the mode value
user_data['elite'].fillna(user_data['elite'].mode()[0], inplace=True)
user_data['friends'].fillna(user_data['friends'].mode()[0], inplace=True)
user_data['name'].fillna(user_data['name'].mode()[0], inplace=True)
print(user_data.isnull().sum())
```

Filled with mode value and we can see that there are no missing values now.

# 4. Suppose you work at Yelp. You have been tasked with a new initiative tocreate a new award for the highest performing food establishmentsrepresented on the Yelp platform. The team at Yelp has determined that anyfood establishment that has both the highest stars rating AND the largesttotal number of reviews (review_count) in their city deserves the Best LocalFood Establishment Award. You have been tasked with writing Python codethat can determine which food establishments in a dataset deserve this newprestigious award.(Important Note: There are some slight variations in some city names like: LasVegas, LasVegas, las vegas. You do not need to worry about cleaning this up before performingyour analysis, this is really a proof-of-concept for your team. Also, some cities will haveno food establishments that deserve to receive this award as it is possible that the foodestablishment with the largest number of reviews does not have the highest rating in thecity.)

```python
# Filtering the data for food establishments
food_data1 = business_data[business_data['categories'].str.contains('Food', na=False)]

# Merging the business data with territory to get state information
merged_data = pd.merge(food_data1, territory, left_on='state', right_on='Abbreviation', how='inner')

# Grouping the data by city and find the food establishment with the highest star rating and the largest number of reviews in each city
best_food_establishments = merged_data.groupby('city').apply(lambda x: x[(x['stars'] == x['stars'].max()) & (x['review_count'] == x['review_count'].max())])

# Reseting the index to remove the multi-level index created by groupby
best_food_establishments = best_food_establishments.reset_index(drop=True)

# Displaying the food establishments deserving the award
print("Food Establishments Deserving the Best Local Food Establishment Award:")
print(best_food_establishments[['name', 'city', 'state', 'stars', 'review_count']])
```

```python
# Setting the plot size
plt.figure(figsize=(12, 6))

# Creating the bar plot
plt.bar(best_food_establishments['city'], best_food_establishments['stars'], color='skyblue')
plt.title('Best Local Food Establishment Award Winners by City')
plt.xlabel('City')
plt.ylabel('Star Rating')
plt.xticks(rotation=45)
plt.ylim(0, 5)

plt.tight_layout()
plt.show()
```

From above representation, the city names are not visible clearly by using matplotlib so I thought of using plotly.express for better visulaization.

```python
import plotly.express as px

# Creating the interactive bar plot
fig = px.bar(best_food_establishments, x='city', y='stars', color='state',
             title='Best Local Food Establishment Award Winners by City',
             labels={'city': 'City', 'stars': 'Star Rating', 'state': 'State'},
             height=500)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

By using plotly.express we can see the Best Local Food Establishment Award Winners by City.

# 5a. Are more reviews posted during certain days of the week? What about certain months of the year? What patterns can you find out about when reviews are posted? Draw a suitable plot.

```python
# Converting the "date" column to a string format
review_data['date'] = review_data['date'].astype(str)

# Filtering out invalid entries in the 'date' column
review_data = review_data[review_data['date'].str.match(r'\d{4}-\d{2}-\d{2}')]

# Converting the "date" column to datetime format
review_data['date'] = pd.to_datetime(review_data['date'])

# Extracting the day of the week and month from the 'date' column
review_data['day_of_week'] = pd.to_datetime(review_data['date']).dt.dayofweek
review_data['month'] = pd.to_datetime(review_data['date']).dt.month

# Counting the number of reviews for each day of the week
day_of_week_counts = review_data['day_of_week'].value_counts().sort_index()

# Plotting the bar plot for distribution of reviews by day of the week
plt.figure(figsize=(10, 6))
plt.bar(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], day_of_week_counts, color='skyblue')
plt.title('Distribution of Reviews by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()

# Counting the number of reviews for each month
month_counts = review_data['month'].value_counts().sort_index()

# Plotting the bar plot for distribution of reviews by month
plt.figure(figsize=(10, 6))
plt.bar(range(1, 13), month_counts, color='salmon')
plt.title('Distribution of Reviews by Month')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.tight_layout()
plt.show()
```

From Visualization, we observed that weekends, particularly Saturdays and Sundays, tend to have a higher volume of reviews compared to weekdays, possibly due to increased leisure time or dining out experiences.

from number of reviews versus month,  we found that holiday seasons, such as May and June coincide with peaks in review activity. These months, typically associated with summer breaks and vacations.

# 5b. Do reviews made on certain days of the week or certain months of theyear tend to be higher/lower rated? What relationship can you findbetween the “stars” rating of a review and when reviews were posted?Draw a suitable plot

```python
# Converting the "date" column to datetime format
review_data['date'] = pd.to_datetime(review_data['date'])

# Extracting the day of the week and month from the "date" column
review_data['day_of_week'] = review_data['date'].dt.day_name()
review_data['month'] = review_data['date'].dt.month_name()

# Aggregate the data to calculate average review ratings by day of the week and by month
avg_ratings_by_day = review_data.groupby('day_of_week')['stars'].mean().reset_index()
avg_ratings_by_month = review_data.groupby('month')['stars'].mean().reset_index()

# Sorting months in chronological order
months_order = list(pd.to_datetime(avg_ratings_by_month['month'], format='%B').dt.month)
avg_ratings_by_month['month'] = pd.Categorical(avg_ratings_by_month['month'], categories=np.array(avg_ratings_by_month['month'])[np.argsort(months_order)])

# Plotting
plt.figure(figsize=(12, 6))

# Bar plot for average review ratings by day of the week
plt.subplot(1, 2, 1)
sns.barplot(x='day_of_week', y='stars', data=avg_ratings_by_day, palette='Set2', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Average Review Ratings by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Review Rating')

# Bar plot for average review ratings by month of the year
plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='stars', data=avg_ratings_by_month, palette='Set2')
plt.title('Average Review Ratings by Month of the Year')
plt.xlabel('Month')
plt.ylabel('Average Review Rating')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

From the graph, we can see very slight change in giving reviewsand rating by dayof the week and Month of the year.
